```python
x = np.linspace(-3, 3, 51)
y_1 = x ** 2
y_2 = 2 * x + 0.5
y_3 = -3 * x - 1.5
y_4 = np.sin(x)
fig, ax = plt.subplots(nrows=2, ncols=2)
ax_1, ax_2, ax_3, ax_4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax_1.plot(x, y_1)
ax_1.set_xlim(-5, 5)
ax_1.set_title(' _1')

ax_2.plot(x, y_2)
ax_2.set_title(' _2')

ax_3.plot(x, y_3)
ax_3.set_title(' _3')

ax_4.plot(x, y_4)
ax_4.set_title(' _4')
```
